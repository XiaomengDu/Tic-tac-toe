```python
# Define the board as a list of empty squares
board = [" " for i in range(9)]

# Define a function to print the board
def print_board():
    print("-------------")
    print("| " + board[0] + " | " + board[1] + " | " + board[2] + " |")
    print("-------------")
    print("| " + board[3] + " | " + board[4] + " | " + board[5] + " |")
    print("-------------")
    print("| " + board[6] + " | " + board[7] + " | " + board[8] + " |")
    print("-------------")

# Define a function to check if someone has won
def check_win(player):
    # Check rows
    for i in range(0, 9, 3):
        if board[i] == board[i+1] == board[i+2] == player:
            return True
    # Check columns
    for i in range(3):
        if board[i] == board[i+3] == board[i+6] == player:
            return True
    # Check diagonals
    if board[0] == board[4] == board[8] == player:
        return True
    if board[2] == board[4] == board[6] == player:
        return True
    return False

# Define the main game loop
def main():
    # Start with player 1
    current_player = "X"

    # Loop until the game is over
    while True:
        # Print the current board
        print_board()

        # Get the current player's move
        while True:
            move = input("Player " + current_player + ", enter your move (1-9): ")
            if move.isdigit() and int(move) in range(1, 10) and board[int(move)-1] == " ":
                break
            else:
                print("Invalid move, try again.")

        # Update the board
        board[int(move)-1] = current_player

        # Check if the game is over
        if check_win(current_player):
            print_board()
            print("Player " + current_player + " wins!")
            break
        elif " " not in board:
            print_board()
            print("Tie game!")
            break

        # Switch to the other player
        current_player = "O" if current_player == "X" else "X"
# Start the game
main()
```
